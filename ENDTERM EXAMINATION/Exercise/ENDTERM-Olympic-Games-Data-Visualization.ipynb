## Age Distribution of Athletes

```python
df_p = df.copy()

plt.figure(facecolor = '#c4d1fb', figsize = (12,6))
sns.histplot(df_p["Athlete_Age"].dropna(), bins = 50, kde = True, color = "#ff33a5")
plt.title("Age Distribution of Olympic Athletes",
          fontfamily = 'Palatino Linotype',
          fontsize = 16,
          fontweight = 'bold')
plt.xlabel("Olympic Athletes' Age",
           fontweight = 'bold')
plt.ylabel("Frequency",
           fontweight = 'bold')
plt.grid(True,
         color = 'gray')
plt.show()
```

The distribution shows that most Olympic athletes are between 20 and 30 years old, 
with a peak around the mid-20s. Very young (below 15) and older athletes (above 40) 
are much less common, indicating that peak performance for most sports occurs in early adulthood.

## Medal Distribution

```python
df_p = df.copy()

plt.figure(facecolor = '#c4d1fb', figsize = (12,6))
sns.countplot(x = "Medal", data = df_p, order = ["Gold", "Silver", "Bronze"], 
              palette = ['#ff33a5', '#ff66c4', '#ff99d8'])
plt.title("Medal Distribution",
          fontfamily = 'Palatino Linotype',
          fontsize = 16,
          fontweight = 'bold')
plt.xlabel("Medal Type",
           fontweight = 'bold')
plt.ylabel("Count",
           fontweight = 'bold')
plt.grid(True,
         color = 'gray')
plt.show()
```

The visualization shows that Bronze medals are the most awarded, followed by Silver and Gold. This is due to some sports awarding two Bronze medals per event, leading to a higher overall count.

## Number of Athletes per Sport

```python
df_p = df.copy()

plt.figure(facecolor = '#c4d1fb', figsize = (12,6))
df_p["Sport"].value_counts().head(15).plot(kind = 'bar', 
            color = '#ff33a5', width = 0.8)
plt.title("Top 15 Sports with Most Athletes",
          fontfamily = 'Palatino Linotype',
          fontsize = 16,
          fontweight = 'bold')
plt.xlabel("Sport",
           fontweight = 'bold')
plt.ylabel("Number of Athletes",
           fontweight = 'bold')
plt.grid(True,
         color = 'gray')
plt.xticks(rotation = 45)
plt.show()
```

The chart shows the top 15 sports with the most athletes, highlighting the dominance of widely practiced and team-based sports like athletics, swimming, and football.

## Athlete Participation Over the Years

```python
df_p = df.copy()

plt.figure(facecolor = '#c4d1fb', figsize = (12,6))
sns.lineplot(data = df_p.groupby("Year")["Athlete_Name"].count().reset_index(), 
             x = "Year", y = "Athlete_Name", marker = "o", color = '#ff33a5')
plt.title("Athlete Participation Over the Years",
          fontfamily = 'Palatino Linotype',
          fontsize = 16,
          fontweight = 'bold')
plt.xlabel("Year",
           fontweight = 'bold')
plt.ylabel("Number of Athletes",
           fontweight = 'bold')
plt.grid(True,
         color = 'gray')
plt.show()
```

The data shows the number of athletes participating in each Olympic year. It highlights fluctuations, with some years having higher participation and others experiencing declines.

## Top 15 Countries by Athlete Count

```python
df_p = df.copy()

plt.figure(facecolor = '#c4d1fb', figsize = (12,6))
df_p["NOC"].value_counts().head(15).plot(kind = 'bar', 
                color = '#ff33a5', width = 0.8)
plt.title("Top 15 Countries with Most Athletes",
          fontfamily = 'Palatino Linotype',
          fontsize = 16,
          fontweight = 'bold')
plt.xlabel("Country (NOC)",
           fontweight = 'bold')
plt.ylabel("Number of Athletes",
           fontweight = 'bold')
plt.grid(True,
         color = 'gray')
plt.xticks(rotation = 45)
plt.show()
```

USA (United States) has the most athletes, followed by FRA (France) and GBR (Great Britain). Other countries, including ITA (Italy), CAN (Canada), GER (Germany), AUS (Australia), JPN (Japan), POL (Poland), SWE (Sweden), URS (Soviet Union), RUS (Russia), HUN (Hungary), CHN (China), and ESP (Spain), have gradually lower numbers, with Spain having the fewest among the top 15.

## Height and Weight Distribution

```python
df_p = df.copy()

plt.figure(facecolor = '#c4d1fb', figsize = (12,6))
sns.scatterplot(data=df_p, x = "Athlete_Height", y = "Athlete_Weight", 
                color = '#ff33a5', alpha = 0.6)
plt.title("Height vs Weight Distribution of Athletes",
          fontfamily = 'Palatino Linotype',
          fontsize = 16,
          fontweight = 'bold')
plt.xlabel("Height (cm)",
           fontweight = 'bold')
plt.ylabel("Weight (Kg)",
           fontweight = 'bold')
plt.grid(True,
         color = 'gray')
plt.show()
```

The scatter plot shows a **positive correlation** between height and weight among athletes, with most data points concentrated between **160-190 cm in height** and **50-100 kg in weight**, while some outliers exceed **200 cm in height** or **160 kg in weight**.

## Gender Distribution

```python
df_p = df.copy()

plt.figure(facecolor = '#c4d1fb', figsize = (12,6))
sns.countplot(x = "Athlete_Sex", data = df_p, palette = ['#ff33a5', '#ff66c4'])
plt.title("Gender Distribution of Olympic Athletes",
          fontfamily = 'Palatino Linotype',
          fontsize = 16,
          fontweight = 'bold')
plt.xlabel("Gender",
           fontweight = 'bold')
plt.ylabel("Count",
           fontweight = 'bold')
plt.grid(True,
         color = 'gray')
plt.show()
```

The gender distribution of Olympic athletes shows that historically, there have been more male athletes than female athletes. However, the gap has gradually narrowed over time as more women have participated in the Games, reflecting increasing gender inclusivity in sports.

## Average Age of Athletes by Year

```python
df_p = df.copy()

df_p['Year'] = pd.to_datetime(df_p['Year'], format='%Y')
df_p_age = df_p.groupby('Year')["Athlete_Age"].mean().reset_index()
print("Average Age of Athletes by Olympic Year")
print(df_p_age.head(30))
```

## Average Athlete Height (cm) by Year

```python
df_height = df_p[['Year', 'Athlete_Height']].dropna()
df_height = df_height.groupby('Year')['Athlete_Height'].mean().reset_index()
print("Average Athlete Height by Year")
print(df_height.head(30))
```

## Total Medals Awarded Overtime

```python
df_p = df.copy()

df_p= df_p[df_p["Medal"].notna()]

medals_per_year = df_p.groupby("Year")["Medal"].count().reset_index()

plt.figure(figsize = (12, 6))
sns.lineplot(data = medals_per_year, x = 'Year', y = 'Medal', marker = 'o', color = '#ff33ab')

plt.title("Overall Trend of Total Medals Awarded Over Time", 
          fontsize = 16, 
          fontweight = 'bold')
plt.xlabel("Olympic Year", 
           fontweight = 'bold')
plt.ylabel("Total Medals Awarded", 
           fontweight = 'bold')
plt.grid(True, linestyle = "--", alpha = 0.5)
plt.show()
```

The trend shows a steady increase in total medals awarded over time, reflecting the growth of the Olympic Games, including more sports, events, and participants.
